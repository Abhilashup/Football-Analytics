# Spain vs Sweden Euro 2020 Waffle Plot for Touches

- The data for this was gathered from StatsBomb via FBref and only certain columns required were selected.

```python
from pywaffle import Waffle
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
from highlight_text import htext, fig_text, HighlightText
```

```python
df = pd.read_csv('spain.csv')
```

```python
df.head()
```

```python
df['Player'] = df['Player'].str.split('\\', expand=True)[0]
```

```python
df.head()
```

```python
#Sorting the data by Touches(from most to least)
df = df.sort_values(by='Touches', ascending=False)
```

```python
df.head()
```

```python
#Dropping the 16 Players row
df = df.drop(16)
```

```python
df
```

```python
#Dropping the touches column as it is longer required and only the areas where the touches were taken in required
df = df.drop('Touches', axis=1)
```

```python
df
```

```python
df = df.T
```

```python
df
```

```python
df.columns = df.iloc[0]
```

```python
df
```

```python
df = df.drop('Player')
```

```python
df
```

```python
df
```
